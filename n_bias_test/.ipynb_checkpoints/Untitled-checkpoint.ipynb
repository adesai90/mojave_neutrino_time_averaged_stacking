```python
################################################################################
# INPUT SECTION
################################################################################


pivot_energy = 100
wrkdir      = "/data/user/adesai/csky/mojave/time_integrated_analysis/n_bias_test"   
index = 2.0  
    

GeV = 1.        # base SkyLab energy units are GeV                        
TeV = 1000*GeV    

E0    = pivot_energy*TeV                # pivot energy 



```

```python
################################################################################
# GET MOJAVE DATA SET (NAME RA DEC av_flux pk_flux)SET Directory
################################################################################

input_file="/data/user/adesai/radio_stacking/mojavexv/1A_wt_rd_flux/MOJAVE_multiyear_complete_sample.txt"


array = np.loadtxt(input_file,dtype='str')
src_name=np.asarray(array[:,0])
src_ra=np.asarray(array[:,1],dtype=float)*np.pi/180
src_dec=np.asarray(array[:,2],dtype=float)*np.pi/180
flux_used=np.asarray(array[:,3],dtype=float)

```

```python
########################################
#  Sample Checks and outdir
########################################

print("---------------------------------------------------------\n")
print("WARNING!!!! CHECK BELOW IF AVG FLUX IS USED (STATEMENT IS TRUE)")
print("Max Avg Flux of sample %s; Flux used:%s"%(21433.22727,np.amax(flux_used)))
if np.amax(flux_used)!=21433.22727:
      print("ERROR: Fluxes not same, Check column number used!!!!")
      #exit() #IN IPYNB THIS WILL KILL THE KERNEL
print("---------------------------------------------------------\n\n\n")


#if len(flux_used)!=no_of_sources:                                                                       
print("---------------------------------------------------------\n")
print("Number of sources used in stacking: %s"%(len(flux_used)))
print("---------------------------------------------------------\n")
no_of_sources=len(flux_used)

```

```python
################################################################################
# Weights!
################################################################################

equal_weights=False

wt=[]
for j in range(len(flux_used)):
    if equal_weights:
        wt.append(1.0)
    else:
        wt.append(flux_used[j])

wt=wt/np.sum(wt)
print("\n WEIGHTS USED FOR THIS RUN:\n %s \n"%(wt))

```

```python
################################################################################
# CSKY SOURCE DESCRIPTION
################################################################################

src = cy.utils.Sources(ra=src_ra,dec=src_dec,weight=wt,deg=False)

flux    = cy.hyp.PowerLawFlux(gamma=index,norm=E0)#,energy_range=(inj_enj_lo,inj_enj_hi))
```

```python


################################################################################
# BACKGROUND
################################################################################

conf = dict(fitter_args = dict(gamma=2.0))#fitter_args = _fmin_method='bfgs')

timer = cy.timing.Timer()
time = timer.time
with time('trial runner construction'):
    tr = cy.get_trial_runner(conf=conf,src=src, ana=ana,flux=flux, sindec_bandwidth=np.radians(.1), mp_cpus=2)

          
```

```python
trial = tr.get_one_trial(n_sig=300, TRUTH=False, seed=np.random.randint(1,1000),poisson=False)

llh = tr.get_one_llh_from_trial(trial)
```

```python
nss = np.linspace(0., 1200., 80)
gammas = np.linspace(1., 4., 80)
llh_space = llh.scan_ts(ns=nss, gamma=gammas)
mesh_ns, mesh_gam = np.meshgrid(nss, gammas)
best_fit_ts = np.max(llh_space[0])
best_fit_ns = mesh_ns.flatten()[np.argmax(llh_space[0].T)]
best_fit_gamma = mesh_gam.flatten()[np.argmax(llh_space[0].T)]
```

```python
import scipy.stats as st
import seaborn as sns
# Calculate critical values
chi2 = st.chi2(2., loc=0., scale=1)
crit_med = chi2.isf(0.5)
crit_90 = chi2.isf(1.0-0.9)
crit_99 = chi2.isf(1.0-0.99)

fig, ax1 = plt.subplots(dpi=150)
ts_space = llh_space[0]

im = ax1.pcolor(mesh_ns, mesh_gam, -1.*(ts_space.T - best_fit_ts), 
                            cmap="Blues_r", vmin=0, vmax=50)#, norm=LogNorm(vmin=0.01, vmax=20))
ax_c = plt.colorbar(im, ax=ax1, format='$%.1f$')
ax_c.ax.tick_params(direction='out')
ax_c.set_label(r"$-2\Delta\,\mathrm{LLH}$")

ax1.contour(mesh_ns, mesh_gam, -1.*(ts_space.T - best_fit_ts), 
                    #levels=[1,4, 9, 16, 25], 
                    levels=[crit_med, crit_90, crit_99], colors="w")
ax1.plot(best_fit_ns, best_fit_gamma, marker="*", color="w", 
                        markersize=10, label='Best-fit')
print(best_fit_ts, best_fit_ns, best_fit_gamma)
ax1.set_ylabel(r"$\gamma$")
ax1.set_xlabel(r"$n_\mathrm{s}$")
ax1.legend(loc=4, facecolor=sns.xkcd_rgb['light grey'])
fig.tight_layout()
plt.savefig("/data/user/adesai/csky/mojave/time_integrated_analysis/n_bias_test/2d_likelihood_plot_gamma_ns_alex_code.png")
```

```python
for i in range(len(mesh_gam)):
    if mesh_gam[i][0]>2.0 and  mesh_gam[i][0]<2.05:
        print(i,mesh_gam[i][0])
        x=mesh_ns[i]
        y=-1.*(ts_space.T[i] - best_fit_ts)
        for j in range(len(y)):
            if y[j]==np.amin(y):
                best_ns=x[j]
        fig, ax1 = plt.subplots(dpi=150)
        ax1.plot(x,y,label="$\gamma$=%.1f"%(2.0))
        ax1.axvline(best_ns, ls = '--', color = 'gray', lw = 1,label='Best Fit $n_\mathrm{s}$=%.1f' % best_ns)
        ax1.set_xlabel(r"$n_\mathrm{s}$")
        ax1.set_ylabel(r"$-2\Delta\,\mathrm{LLH}$")
        ax1.legend()
        fig.tight_layout()
        plt.savefig("/data/user/adesai/csky/mojave/time_integrated_analysis/n_bias_test/2d_likelihood_plot_gamma_%s.png"%(2.0))
```
